```python
import numpy as np
import pandas as pd

np.random.seed(0)
df=pd.DataFrame({'Sex': np.random.randint(2, size=1000), 'Height': np.random.randint(low=100, high=250, size=1000)})
df.head(10)
```

```python
df.iloc[0]['Height']
```

```python
def loss_regression(y_predicted, y_true):
    return (y_predicted-y_true)**2
```

```python
df['Predicted_height']=df.iloc[0]['Height']
df.head(10)
```

```python
df['Fun_loss']=df[['Height', 'Predicted_height']].apply(lambda x: loss_regression(x[1], x[0]), axis=1)
```

```python
df.head()
```

```python
def empirical_risk(loss):
    return sum(loss)/len(loss)
```

```python
df['Fun_loss']
```

```python
empirical_risk(df['Fun_loss'].values)
```
