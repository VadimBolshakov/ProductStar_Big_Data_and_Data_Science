```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('productstar_df.csv')
df= df[df['order_state'] == 'complete']
df['completed_at']= pd.to_datetime(df['completed_at'])
df=df[df['completed_at'] >= '2018-01-01']
df
```

```python
df['date']=df['completed_at'].dt.date
df
```

```python
df_t = df.groupby(['user_id', 'date'], as_index=False).count()[['user_id', 'date']]
df_t
```

```python
users=pd.DataFrame(df_t['user_id'].unique(), columns=['user_id'])
users['u_num']=users.index
users
```

```python
dates=pd.DataFrame(df_t['date'].unique(), columns=['date'])
dates['d_num']=dates.index
dates
```

```python
df_t=df_t.merge(users, on='user_id').merge(dates, on='date')
```

```python
arr=np.zeros((len(users), len(dates)))
arr
```

```python
arr[df_t['u_num'], df_t['d_num']]=1
arr
```

```python
rss=[]
for i in range(len(users)):
    if i % 1000 == 0:
        print(i)
    rs=np.dot(arr, arr[i])
    rs[i] = 0
    rss.append((i,np.where(rs == rs.max())[0][0]))
```

```python
pd.DataFrame(rss)
```
